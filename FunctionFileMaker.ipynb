```python
#general utility functions

#path for creating functions in the data pack, replace it by "" if you want it to create the function in the same file
pathstring="SMTDataPack/data/utility/function/"

#number of sheeps
N=144 

#creates a function that summons N sheeps to the same position and names each with a number
filename =pathstring+ "multisetup.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    for i in range(N):
        f.write(f"summon minecraft:sheep ~ ~ ~ {{CustomName:'\"{i}\"'}}\n")

#puts N sheep into a team, so they don't collide with each other
filename = pathstring+"nocollteam.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("team add noclip\n")
    f.write("team modify noclip collisionRule never\n")
    for i in range(N):
        f.write(f"team join noclip @e[type=minecraft:sheep,name={i}]\n")


#creates a function that sets up the necessary datastructures for position and/or direction tracking
filename = pathstring+"tracksetup.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("scoreboard objectives add x dummy\n")
    f.write("scoreboard objectives add y dummy\n")
    f.write("scoreboard objectives add z dummy\n")
    f.write("scoreboard objectives add yaw dummy\n")
    f.write("scoreboard objectives add pitch dummy\n")


#creates a function that prints the name, the position and the direction of all sheep
pozmult=1 #multiplies the positon values, in order to log data more precisly
rotmult=1 #multiplies the rotation values, in order to log data more precisly
filename = pathstring+"trackpozandrot.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write(f"execute as @e[type=sheep] store result score @s x run data get entity @s Pos[0] {pozmult}\n")
    f.write(f"execute as @e[type=sheep] store result score @s y run data get entity @s Pos[1] {pozmult}\n")
    f.write(f"execute as @e[type=sheep] store result score @s z run data get entity @s Pos[2] {pozmult}\n")
    f.write(f"execute as @e[type=sheep] store result score @s yaw run data get entity @s Rotation[0] {rotmult}\n")
    f.write(f"execute as @e[type=sheep] store result score @s pitch run data get entity @s Rotation[1] {rotmult}\n")
    #comment if you dont need position
    f.write("execute as @e[type=minecraft:sheep] run tellraw @a [{\"selector\":\"@s\"},{\"text\":\" | Pos: \"},{\"score\":{\"name\":\"@s\",\"objective\":\"x\"}},{\"text\":\", \"}, {\"score\":{\"name\":\"@s\",\"objective\":\"y\"}},{\"text\":\", \"},{\"score\":{\"name\":\"@s\",\"objective\":\"z\"}}]\n")
    #comment if you dont need direction
    f.write("execute as @e[type=minecraft:sheep] run tellraw @a [{\"selector\":\"@s\"}, {\"text\":\" | Facing: \"}, {\"score\":{\"name\":\"@s\",\"objective\":\"yaw\"}},{\"text\":\", \"}, {\"score\":{\"name\":\"@s\",\"objective\":\"pitch\"}}]\n")

```

```python
#path for creating functions in the data pack, replace it by "" if you want it to create the function in the same file
pathstring="SMTDataPack/data/smt/function/"

N=10 #number os sheep in a single iteration
loop=10 #number of iterations per site

maxTick=100 #waiting time before discarding measurment if none of the sheep moved
dy=10 #vertical separation

#postion of the structure, change at will
x0=0
z0=0

#Creation of the functions needed for single measurment
blocknumber=100 #number of blocks in the field(for data logging)

filename = pathstring+"singlemeasure.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("scoreboard players add #tick vars 1\n\n")
    f.write("execute as @e[type=minecraft:sheep] at @s unless entity @e[type=minecraft:armor_stand, name=SheepMarker, distance=..0.0001] run scoreboard players remove #Nsheep vars 1\n")
    f.write(f"execute as @e[type=minecraft:sheep] at @s unless entity @e[type=minecraft:armor_stand, name=SheepMarker, distance=..0.0001] run tellraw @p [\"Target found({blocknumber} block) at tick: \",{{\"score\":{{\"name\":\"#myVar\",\"objective\":\"timer\"}} }}]\n")
    f.write("execute as @e[type=minecraft:sheep] at @s unless entity @e[type=minecraft:armor_stand, name=SheepMarker, distance=..0.0001] run kill @s\n")
    f.write("kill @e[type= minecraft:item]\n\n")
    f.write("execute if score #Nsheep vars matches 0 run setblock ~ ~-1 ~ minecraft:air\n")
    f.write("execute if score #Nsheep vars matches 0 if score @p vars matches 1.. run function smt:restart\n")

filename = pathstring+"singlesetup.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("scoreboard objectives add vars dummy \"Variables\"\n")
    f.write("scoreboard players reset * vars\n")
    f.write("scoreboard players set #tick vars 0\n")
    f.write("scoreboard players set @p vars 0\n")
    f.write("scoreboard players set #Nsheep vars 0\n")
    f.write("scoreboard players set #poz vars 0\n\n")
    f.write(f"fill ~-2 ~10 ~-2 ~2 ~10 ~2 minecraft:oak_leaves[persistent=true]\n")
    f.write(f"fill ~ ~10 ~ ~ ~10 ~ minecraft:cherry_leaves[persistent=true]\n")
    f.write("kill @e[type=minecraft:armor_stand]\n")
    f.write("kill @e[type= minecraft:sheep]\n")
    f.write("kill @e[type= minecraft:item]\n\n")
    f.write(f"summon minecraft:armor_stand ~ ~9.5 ~ {{Marker:1b,NoGravity:1b,CustomName:\"\\\"SheepMarker\\\"\",PersistenceRequired:1b}}\n")

#code that generates the function files for the SMT
filename = pathstring+"spawn.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write(f"tp @e[type=minecraft:armor_stand,name=SheepMarker] ~ ~{dy-0.5} ~\n\n")
    f.write("kill @e[type= minecraft:sheep]\n")
    f.write("kill @e[type= minecraft:item]\n\n")
    f.write(f"scoreboard players set #Nsheep vars {N}\n")
    f.write("scoreboard players set #tick vars 0\n\n")
    f.write("team add noclip\n")
    f.write("team modify noclip collisionRule never\n")
    for i in range(N):
        f.write(f"summon minecraft:sheep ~ ~{dy-0.5} ~ {{CustomName:'\"{i}\"'}}\n")
        f.write(f"team join noclip @e[type=minecraft:sheep,name={i}]\n")



filename = pathstring+"measure.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("scoreboard players add #tick vars 1\n\n")
    f.write("execute as @e[type=minecraft:sheep] at @s unless entity @e[type=minecraft:armor_stand, name=SheepMarker, distance=..0.0001] run scoreboard players remove #Nsheep vars 1\n")
    f.write("execute as @e[type=minecraft:sheep] at @s unless entity @e[type=minecraft:armor_stand, name=SheepMarker, distance=..0.0001] run tellraw @p [\"poz: \",{\"score\":{\"name\":\"#poz\",\"objective\":\"vars\"} },\"tick: \",{\"score\":{\"name\":\"#tick\",\"objective\":\"vars\"} }]\n")
    f.write("execute as @e[type=minecraft:sheep] at @s unless entity @e[type=minecraft:armor_stand, name=SheepMarker, distance=..0.0001] run kill @s\n")
    f.write("kill @e[type= minecraft:item]\n\n")
    f.write(f"execute if score #Nsheep vars matches {N} if score #tick vars matches {maxTick}.. run function smt:break\n\n")
    f.write("execute if score #Nsheep vars matches 0 run setblock ~ ~-1 ~ minecraft:air\n")
    f.write("execute if score #Nsheep vars matches 0 if score @p vars matches 1.. run function smt:restart\n")


filename = pathstring+"restart.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("function smt:spawn\n")
    f.write("setblock ~ ~-1 ~ minecraft:redstone_block\n")
    f.write("scoreboard players remove @p vars 1\n")

filename = pathstring+"break.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("scoreboard players set #Nsheep vars 0\n")
    f.write("scoreboard players set @p vars 0\n")
    f.write("kill @e[type= minecraft:sheep]\n")
    f.write("kill @e[type= minecraft:item]\n")
    f.write("tellraw @p [\"poz: \",{\"score\":{\"name\":\"#poz\",\"objective\":\"vars\"} },\"tick:0.0 \"]\n")

filename = pathstring+"draw.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("fill ~ ~-2 ~ ~ ~-16 ~ minecraft:lime_concrete\n")

filename = pathstring+"erase.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("fill ~ ~-2 ~ ~ ~-16 ~ minecraft:air\n")

filename = pathstring+"loop.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("execute if score #Nsheep vars matches 0 if score @p vars matches 0 run function smt:nextswep\n")

filename = pathstring+"setup.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write("scoreboard objectives add vars dummy \"Variables\"\n")
    f.write("scoreboard players reset * vars\n")
    f.write("scoreboard players set #tick vars 0\n")
    f.write("scoreboard players set @p vars 0\n")
    f.write("scoreboard players set #Nsheep vars 0\n")
    f.write("scoreboard players set #poz vars 0\n\n")
    f.write(f"fill {x0-2} -39 {z0-2} {x0+16} -14 {z0+16} minecraft:air\n")
    f.write("kill @e[type=minecraft:armor_stand]\n")
    f.write("kill @e[type= minecraft:sheep]\n")
    f.write("kill @e[type= minecraft:item]\n\n")
    f.write(f"fill {x0} -39 {z0} {x0+5} -39 {z0+5} minecraft:redstone_block\n")
    f.write(f"fill {x0-1} -29 {z0-1} {x0+8} -29 {z0+8} minecraft:oak_leaves[persistent=true]\n")
    f.write(f"fill {x0} -29 {z0} {x0+5} -29 {z0+5} minecraft:cherry_leaves[persistent=true]\n")
    f.write(f"fill {x0+9} -21 {z0+9} {x0+16} -37 {z0+16} minecraft:spruce_leaves[persistent=true]\n")
    f.write(f"fill {x0+10} -21 {z0+10} {x0+15} -36 {z0+15} minecraft:air\n")
    f.write(f"fill {x0+10} -21 {z0+10} {x0+15} -21 {z0+15} minecraft:green_stained_glass\n")
    f.write(f"fill {x0+1} -14 {z0+1} {x0+4} -14 {z0+4} minecraft:green_stained_glass\n")
    f.write(f"summon minecraft:armor_stand {x0} -28 {z0} {{Marker:1b,NoGravity:1b,CustomName:\"\\\"SheepMarker\\\"\",PersistenceRequired:1b}}\n")

filename = pathstring+"nextswep.mcfunction"
with open(filename, "w", encoding="utf-8") as f:
    f.write(f"fill {x0} -39 {z0} {x0+5} -38 {z0+5} minecraft:air\n")
    f.write(f"scoreboard players set @p vars {loop}\n\n")
    poz=0
    for i in range(6):
        for j in range(6):
            f.write(f"execute if score #poz vars matches {poz} run setblock {x0+i} -38 {z0+j} repeating_command_block[conditional=false,facing=up]{{Command:\"function smt:measure\",auto:0b}}\n")
            poz+=1

    f.write("execute if score #poz vars matches 36.. run setblock ~ ~1 ~ minecraft:air\n\n")
    f.write("execute if score #poz vars matches 36.. run scoreboard players set @p vars 0\n")
    f.write("scoreboard players set #Nsheep vars 0\n")
    f.write(f"fill {x0} -39 {z0} {x0+5} -39 {z0+5} minecraft:redstone_block\n")
    f.write("scoreboard players add #poz vars 1\n")
    
```
